```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the CSV File
df = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

```python
# Function to identify which color to be used to plot 
def pltcolor(dataf):
    '''
    Function to identify which color to be used to plot based on gender
    INPUT: dataframe
    OUTPUT: Updated Dataframe with color column added
    '''
    cols = []
    lst = dataf['sex']
    #Loop thru the gender list and set colors accoringly
    for l in lst:
        if l == 'male':
            cols.append('red')
        elif l=='female':
            cols.append('green')
        else:
            cols.append('blue')
    #appends the plot colors to the data frame
    dataf['plotcolors'] = cols
    return dataf 

# Colors to be used for plotting
cols = pltcolor(df)['plotcolors']
# Get the array of fare paid
farepaid = df['fare']
# get the array of age
age = df['age']

plt.scatter(farepaid, age,c=cols)
plt.xlabel('Fare Paid')
plt.ylabel('Age')
plt.title('Scatter plot with the fair paid and the age, differed by plot color by gender')
plt.show()
```
